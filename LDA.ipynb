```python
#return a text with no punctuations in it
import regex
from nltk.tokenize import RegexpTokenizer
def remove_punctuations(text):
    text=regex.sub('\n'," ",text)
    text=regex.sub('\xe2'," ",text)
    text=regex.sub('\u2019'," ",text)
    
    tokenizer=RegexpTokenizer(r'\w+') #removes punctuations
    text=tokenizer.tokenize(text)
    text=[token.lower() for token in text]
    return " ".join(text)
```

```python
dataset['filtered_corpus']=dataset['text'].apply(lambda x: remove_punctuations(str(x)))
```

```python
import nltk
from nltk.tokenize import word_tokenize
def add_bigrams(text):
    tokens=nltk.word_tokenize(text)
    bi_grams=nltk.bigrams(tokens)
    bi_grams_list=[]
    for i in bi_grams:
        bi_grams_list.append(" ".join(i))
    return bi_grams_list
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize #word tokenize will convert sentence into list of words
# this function return text which has no stop words
def remove_stopwords(text):
    stop_words=set(stopwords.words('english'))
    tokens=word_tokenize(text) # tokens will be list of words
    filtered_sentence=[word for word in tokens if not word in stop_words] # 
    return " ".join(filtered_sentence)
```

```python
from nltk.tokenize import word_tokenize
def generate_corpus_for_lda(text):
    #try:
    text=remove_punctuations(text)
    text=remove_stopwords(text) #upto this point string doesnt contain stopwords
   # except:
    #    "encoding error"
    try:
        mono_grams=word_tokenize(text)
    except:
        if text is None:
            print text
            print type(text)
        print " error"
    bi_grams=add_bigrams(text)
    return mono_grams+bi_grams # will output a list of monograms and bigrams
    
```

```python
dataset['corpus_for_lda']=dataset['filtered_corpus'].apply(lambda x: generate_corpus_for_lda(str(x)))
```

```python
from gensim import corpora
dictionary=corpora.Dictionary(temp_list)
training_corpus=[dictionary.doc2bow(doc) for doc in list(dataset['corpus_for_lda'])]
corpora.MmCorpus.serialize('got_doc.mm',training_corpus)
mm_training_corpus=gensim.corpora.MmCorpus('got_doc.mm') # this is our input corpus for LDA
print(mm_training_corpus)
```

```python
dictionary.save_as_text(fname='dic_got_lda.txt', sort_by_word=True)
id2word = gensim.corpora.Dictionary.load_from_text('dic_got_lda.txt')
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=mm_training_corpus, id2word=id2word, num_topics=20, update_every=0, chunksize=10000, passes=200)
#num_topic --> no of topics you want in lda
#passes --> Number of iterations
```
